```python
import os
import pandas as pd
```

```python
listdir = os.listdir('./202101/')
```

```python
df = pd.DataFrame()

count = 0

for name in listdir:
    temp = pd.read_excel('./202101/{}'.format(name)) ################
    df = pd.concat([df, temp])
    count += len(df)
```

```python
df = df.sort_values(by=['일자','시도(출발)','시군구(출발)'])
df = df.reset_index(drop=True)
```
